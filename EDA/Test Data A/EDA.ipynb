3.3. check the missing values

```python
row_count = len(data)  # get total number of rows to compute percentages
missing_count = data.isna().sum()  # count missing (NaN) values for each column
missing_percent = (missing_count / row_count * 100) if row_count > 0 else (missing_count * 0)  # compute percent per column, handle empty DF

# assemble the summary table with required columns
missing_table = pd.DataFrame({
    "column_name": missing_count.index,          # column names
    "missing_count": missing_count.values,       # number of missing values
    "missing_percent": missing_percent.values    # percentage of missing values
})

missing_table["missing_percent"] = missing_table["missing_percent"].round(2)  # round percentages to 2 decimals for readability
missing_table = missing_table.sort_values("missing_count", ascending=False).reset_index(drop=True)  # sort by most missing first

print(missing_table)  # display the table
```

3.4. check the duplicate values

```python
n_rows = len(data)  # total number of rows

dup_mask = data.duplicated(keep=False)  # True for every row that belongs to a duplicate group (all members)
rows_in_dup_groups = dup_mask.sum()  # count of rows that are part of any duplicate group (includes the first occurrences)
dup_percent = (rows_in_dup_groups / n_rows * 100) if n_rows > 0 else 0.0  # percentage of rows in duplicate groups

dup_rows = data[dup_mask].copy()  # all rows that are duplicates 

# Build a compact summary of duplicate groups
# dropna=False ensures rows with NaNs are grouped properly rather than ignored
duplicate_groups = (
    data[dup_mask]                                  # keep only rows that are part of duplicate groups
    .groupby(list(data.columns), dropna=False)      # group by all columns to identify identical rows
    .size()                                         # count occurrences per identical row
    .reset_index(name="count")                      # move the count into a column
    .sort_values("count", ascending=False)          # show the most frequent duplicates first
    .reset_index(drop=True)                         # clean index for neat display
)

# number of "extra" duplicates beyond the first instance of each group
extra_duplicate_rows = data.duplicated(keep="first").sum()  # counts only the second+ occurrences

print(f"Rows in duplicate groups: {rows_in_dup_groups} of {n_rows} ({dup_percent:.2f}%)")  # quick summary
print(f"Extra duplicate rows beyond first occurrences: {extra_duplicate_rows}")  # additional metric
print("\nDuplicate groups (each unique duplicated row with its frequency):")
print(duplicate_groups.head(20))  # preview top 20 duplicate groups
```

4.1. density plots for numerical features

```python
import numpy as np
import math                       # for ceiling when computing rows
import pandas as pd               # for numeric coercion / NA handling
import matplotlib.pyplot as plt   # plotting
import seaborn as sns             # KDE plots

# --- columns to plot (keep only those that actually exist in `data`) ---
cols = ['ankle','arm-length','bicep','calf','chest','forearm','hip',
        'leg-length','shoulder-breadth','shoulder-to-crotch','thigh','waist',
        'wrist','height_cm','weight_kg']
cols = [c for c in cols if c in data.columns]  # drop any names not present

# --- layout: choose number of columns per row, compute how many rows we need 
plots_per_row = 4                               # 4 columns per row (gives 4x4 for 16 vars)
n = len(cols)                                    # how many variables to plot
nrows = max(1, math.ceil(n / plots_per_row))     # rows needed to fit all plots
ncols = plots_per_row                            # columns per row

# create the subplot grid sized to fit everything 
fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 4, nrows * 3))  # scale figure size
axes = axes.flatten() if isinstance(axes, (list, np.ndarray)) else [axes]  # flatten axes for easy indexing

# loop over columns and draw KDE for each
for i, col in enumerate(cols):
    s = pd.to_numeric(data[col], errors='coerce')      # coerce to numeric so KDE wont choke on strings
    s = s.dropna()                                     # drop NaNs before density estimation
    ax = axes[i]                                       # pick the corresponding subplot axis
    if len(s) >= 2:                                    # need at least 2 points for KDE
        sns.kdeplot(s, fill=True, ax=ax)               # draw the density curve
    else:
        ax.text(0.5, 0.5, "Not enough data", ha='center', va='center')  # informative placeholder
    ax.set_title(f'Density Plot of {col}')             # title for this subplot
    ax.set_xlabel(col)                                 # x-axis label
    ax.set_ylabel('Density')                           # y-axis label

# --- clean up any leftover (unused) axes if grid > number of plots ---
for j in range(len(cols), nrows * ncols):
    fig.delaxes(axes[j])                                # remove empty subplots

plt.tight_layout()                                      # tidy spacing
plt.savefig("distribution.png", dpi=300, bbox_inches="tight")
plt.show()                                              # render the figure
```

5. Bivariate data visualization

5.1. bar charts comparing each numeric variable by weight_category

```python
import math                            # for computing grid size
import pandas as pd                    # for grouping/aggregation
import numpy as np                     # for numeric ops / SEM
import matplotlib.pyplot as plt        # for plotting

# --- numeric columns to compare against weight_category ---
cols = ['ankle','arm-length','bicep','calf','chest','forearm','hip',
        'leg-length','shoulder-breadth','shoulder-to-crotch','thigh','waist',
        'wrist','height_cm']

# keep only columns present in the DataFrame to avoid KeyErrors
cols = [c for c in cols if c in data.columns]

# --- enforce desired category order: low  normal  high (NaN last if present) ---
order = ["low weight", "normal weight", "high weight"]
cat_dtype = pd.CategoricalDtype(categories=order, ordered=True)
ser = data["weight_category"].astype(cat_dtype)                 # keep NaNs as NaN
has_nan = ser.isna().any()
cat_order = order + ([np.nan] if has_nan else [])               # NaN group last if it exists
xtick_labels = [("NaN" if pd.isna(x) else str(x)) for x in cat_order]

# working frame: categorical weight_category + numeric columns coerced to numeric
work = pd.DataFrame({"weight_category": ser})
for c in cols:
    work[c] = pd.to_numeric(data[c], errors="coerce")           # coerce non-numeric to NaN

# helper: compute mean and SEM aligned to cat_order
def mean_sem(df, var):
    # observed=False keeps ALL categories (even if unobserved) and silences the FutureWarning
    grp = df.groupby("weight_category", dropna=False, observed=False)[var]
    means = grp.mean()                                          # mean per category
    counts = grp.count()                                        # non-NaN counts for var
    stds = grp.std()                                            # std per category
    sems = stds / np.sqrt(counts.replace(0, np.nan))            # SEM = std / sqrt(n)
    return means.reindex(cat_order), sems.reindex(cat_order)    # align to fixed order

# --- layout: TWO plots per row ---
plots_per_row = 2                                               # two per row
n = len(cols)                                                   # number of variables
nrows = max(1, math.ceil(n / plots_per_row))                    # required rows
ncols = plots_per_row                                           # columns per row

# figure sized for two-wide layout (wider/taller per plot for readability)
fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 5.2, nrows * 3.8))
axes = np.array(axes).reshape(-1) if isinstance(axes, np.ndarray) else np.array([axes])

# loop over variables and draw bar charts with error bars
for i, var in enumerate(cols):
    ax = axes[i]                                                # pick subplot axis
    means, sems = mean_sem(work, var)                           # compute stats
    x = np.arange(len(cat_order))                               # x positions

    # Use 0.0 for missing means so bars render; annotations will skip NaNs
    heights = np.nan_to_num(means.values, nan=0.0)
    yerr = sems.values

    bars = ax.bar(x, heights, yerr=yerr, capsize=4)             # bars + SEM

    # annotate bar tops with mean values (skip NaNs)
    for rect, val in zip(bars, means.values):
        if pd.notna(val):
            ax.text(rect.get_x() + rect.get_width()/2,
                    rect.get_height(),
                    f"{val:.2f}",
                    ha="center", va="bottom", fontsize=8)

    # cosmetics
    ax.set_title(f"{var} by Weight Category")
    ax.set_xlabel("Weight Category")
    ax.set_ylabel("Mean")
    ax.set_xticks(x)
    ax.set_xticklabels(xtick_labels, rotation=45, ha='right')

# remove any unused axes (when grid > number of variables)
for j in range(len(cols), nrows * ncols):
    fig.delaxes(axes[j])

plt.tight_layout()                                              # tidy spacing
plt.show()                                                      # display
fig.savefig("weight category vs numerical columns.png", dpi=300, bbox_inches="tight")
```

* Measurements that Increase with Weight
    * Waist circumference demonstrates the most pronounced progression, rising from approximately 73 cm in low-weight individuals to around 113 cm in high-weight individuals.
    * Chest, hip, thigh, bicep, calf, and forearm measurements all show gradual increases across weight categories
    * Wrist size also increases slightly with weight category
* Measurements that Remain Stable
    * Height displays little variation across weight categories (approximately 170-175 cm), which is expected since height is independent of weight.
    * Arm length, leg length, and shoulder breadth remain fairly consistent across all three groups.
    * Shoulder-to-crotch length shows minimal change

5.2. bar charts comparing each numeric variable by gender

```python
import math                            # for computing grid size
import pandas as pd                    # for grouping/aggregation
import numpy as np                     # for numeric ops / SEM
import matplotlib.pyplot as plt        # for plotting

# --- numeric columns to compare against gender ---
cols = ['ankle','arm-length','bicep','calf','chest','forearm','hip',
        'leg-length','shoulder-breadth','shoulder-to-crotch','thigh','waist',
        'wrist','height_cm','weight_kg']

# keep only columns present in the DataFrame to avoid KeyErrors
cols = [c for c in cols if c in data.columns]

# canonicalize category labels; show NaN as a literal "NaN" category label for plotting
cat_series = data['gender'].astype('object')        # ensure object dtype for mixed labels
categories = cat_series.fillna("NaN")               # replace NaN with string label

# working frame: category + numeric columns coerced to numeric
work = pd.DataFrame({'gender': categories})
for c in cols:
    work[c] = pd.to_numeric(data[c], errors='coerce')  # coerce non-numeric to NaN

# choose category order by frequency (most common first)
cat_order = (work['gender']
             .value_counts(dropna=False)
             .index.tolist())

# helper: compute mean and SEM aligned to cat_order
def mean_sem(df, var):
    grp = df.groupby('gender', dropna=False)[var]       # group by category
    means = grp.mean()                                  # mean per category
    counts = grp.count()                                # non-NaN counts
    stds = grp.std()                                    # std per category
    sems = stds / np.sqrt(counts.replace(0, np.nan))    # SEM = std / sqrt(n)
    return means.reindex(cat_order), sems.reindex(cat_order)

# --- layout: TWO plots per row ---
plots_per_row = 2
n = len(cols)
nrows = max(1, math.ceil(n / plots_per_row))
ncols = plots_per_row

# figure sized for two-wide layout (roomy per-plot size)
fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 5.2, nrows * 3.8))
axes = np.array(axes).reshape(-1) if isinstance(axes, np.ndarray) else np.array([axes])

# loop over variables and draw bar charts with error bars
for i, var in enumerate(cols):
    ax = axes[i]
    means, sems = mean_sem(work, var)
    x = np.arange(len(cat_order))
    bars = ax.bar(x, means.values, yerr=sems.values, capsize=4)

    # annotate bar tops with mean values (skip NaNs)
    for rect, val in zip(bars, means.values):
        if pd.notna(val):
            ax.text(rect.get_x() + rect.get_width()/2,
                    rect.get_height(),
                    f"{val:.2f}",
                    ha="center", va="bottom", fontsize=8)

    # cosmetics
    ax.set_title(f"{var} by Gender")
    ax.set_xlabel("Gender")
    ax.set_ylabel("Mean")
    ax.set_xticks(x)
    ax.set_xticklabels(cat_order, rotation=45, ha='right')

# remove any unused axes (when grid > number of variables)
for j in range(len(cols), nrows * ncols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# save the figure
fig.savefig("gender vs numerical columns.png", dpi=300, bbox_inches="tight")
```

* Measurements Where Males are Larger
    * Height shows males averaging approximately 175 cm compared to females at around 162 cm
    * Shoulder breadth is noticeably larger in males (approximately 38 cm) than females (around 34 cm)
    * Chest circumference is greater in males (about 105 cm) compared to females (approximately 95 cm)
    * Arm length and leg length are both longer in males than females
    * Shoulder-to-crotch length is greater in males (around 67 cm) versus females (approximately 60 cm)
    * Wrist size is slightly larger in males than females
* Measurements Where Females are Larger
    * Hip circumference is notably larger in females (approximately 105 cm) compared to males (around 98 cm)
    * Thigh circumference is greater in females (around 56 cm) than males (approximately 53 cm)
    * Weight shows females averaging slightly higher (around 70 kg) than males (approximately 65 kg) in this dataset

6.1. correlation analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# columns of interest
cols = ['ankle','arm-length','bicep','calf','chest','forearm','hip',
        'leg-length','shoulder-breadth','shoulder-to-crotch','thigh','waist',
        'wrist','height_cm','weight_kg']

# keep existing columns and coerce to numeric
corr_data = data[[c for c in cols if c in data.columns]].apply(pd.to_numeric, errors='coerce')

# drop unusable columns (all-NaN or constant)
corr_data = corr_data.dropna(axis=1, how='all')
corr_data = corr_data.loc[:, corr_data.nunique(dropna=True) > 1]

# compute Pearson correlation
corr = corr_data.corr(method='pearson', min_periods=2)

# plot heatmap (no mask, no printing)
fig, ax = plt.subplots(figsize=(max(8, 0.6*len(corr.columns)),
                                max(6, 0.6*len(corr.columns))))
sns.heatmap(
    corr,
    annot=True, fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5, linecolor="white",
    square=True,
    cbar_kws={"shrink": 0.8}
)
ax.set_title("Correlation Matrix for Selected Measurements")
fig.tight_layout()

# save before show (adjust filename/path as you like)
fig.savefig("correlation heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

* Strong relationships with height:
    * Leg-length (0.88), arm-length (0.90), and shoulder-to-crotch (0.86) have very strong positive correlations with height. This makes sense because taller people naturally have longer limbs and torsos.
    * Height also shows a moderate positive correlation with weight (0.57), meaning taller people tend to weigh more.
* Moderate relationships with height:
    * Shoulder-breadth (0.69), ankle (0.68), and wrist (0.64) show moderate positive correlations with height. These measurements tend to increase somewhat as height increases.
* Weaker relationships with height:
    * Bicep (0.42), calf (0.36), waist (0.33), chest (0.31), and hip (0.32) show weak to moderate correlations with height

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- numeric columns to compare (one removed already) ---
cols = ['ankle','arm-length','bicep','calf','chest','forearm','hip',
        'leg-length','shoulder-breadth','shoulder-to-crotch','thigh','waist',
        'wrist','height_cm']
cols = [c for c in cols if c in data.columns]  # keep only existing columns

# --- categorical columns ---
# Gender: keep original labels, show missing as literal "NaN"
genders = data['gender'].astype('object').fillna("NaN")

# Weight category: enforce fixed order low  normal  high, keep real NaNs as NaN
weight_order_base = ["low weight", "normal weight", "high weight"]
cat_dtype = pd.CategoricalDtype(categories=weight_order_base, ordered=True)
weights_cat = data['weight_category'].astype(cat_dtype)  # NaNs remain NaN
has_nan_weight = weights_cat.isna().any()
weight_order = weight_order_base + ([np.nan] if has_nan_weight else [])  # NaN last if present

# --- working frame: categories + numeric columns coerced to numeric ---
work = pd.DataFrame({'gender': genders, 'weight_category': weights_cat})
for c in cols:
    work[c] = pd.to_numeric(data[c], errors='coerce')

# --- category order for gender (by frequency; customize if you prefer a fixed order) ---
gender_order = work['gender'].value_counts(dropna=False).index.tolist()

# --- layout: TWO plots per row ---
plots_per_row = 2
n = len(cols)
nrows = max(1, math.ceil(n / plots_per_row))
ncols = plots_per_row

fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 6.0, nrows * 4.2))
axes = np.array(axes, ndmin=1).ravel()  # flatten safely for n=1 as well

# --- bar grouping settings ---
group_width = 0.8                          # total width for each gender group
k = max(1, len(weight_order))              # bars per group (one per weight category)
bar_width = group_width / k
x_pos = np.arange(len(gender_order))       # positions for gender groups

# --- plot each numeric variable ---
for i, var in enumerate(cols):
    ax = axes[i]

    # mean per (gender, weight_category), aligned to chosen order
    grouped = (
        work.groupby(['gender', 'weight_category'], dropna=False, observed=False)[var]
            .mean()
            .unstack('weight_category')
    )
    grouped = grouped.reindex(index=gender_order, columns=weight_order)

    # grouped bars: one color per weight_category (NaN column included last if present)
    for j, wc in enumerate(weight_order):
        y = grouped[wc].values
        offsets = x_pos - group_width/2 + j * bar_width + bar_width/2
        bars = ax.bar(offsets, y, width=bar_width, label=("NaN" if pd.isna(wc) else str(wc)))

        # annotate bar tops with mean values (skip NaNs)
        for rect, val in zip(bars, y):
            if pd.notna(val):
                ax.text(rect.get_x() + rect.get_width()/2,
                        rect.get_height(),
                        f"{val:.2f}",
                        ha="center", va="bottom", fontsize=8)

    # cosmetics
    ax.set_title(f"{var} by Gender and Weight Category")
    ax.set_xlabel("Gender")
    ax.set_ylabel("Mean")
    ax.set_xticks(x_pos)
    ax.set_xticklabels(gender_order)
    ax.margins(x=0.02)

# remove any unused axes if grid > number of variables
for j in range(len(cols), nrows * ncols):
    fig.delaxes(axes[j])

# single shared legend (top center)
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title="Weight Category",
           loc="upper center", ncol=min(4, len(weight_order)), frameon=False)

plt.tight_layout(rect=(0, 0, 1, 0.93))  # leave space for the legend

# save before show
fig.savefig("gender_vs_numericals_by_weight_category.png", dpi=300, bbox_inches="tight")
plt.show()
```

* Overall structure
    * Each chart compares measurements between males and females, with three weight categories shown for each gender: low weight (blue bars), normal weight (orange bars), and high weight (green bars).
* Key Patterns
    * Gender differences: Males generally show larger measurements than females for most body parts, including chest, waist, shoulders, and hips. However, some measurements like thigh and waist show smaller differences between genders.
    * Weight category effects: High weight individuals (green bars) consistently demonstrate larger measurements across almost all body parts compared to normal and low weight individuals. This pattern is evident in measurements like waist, hip, chest, and thigh.
* General trend
    * For each body part, the pattern typically follows: high weight > normal weight > low weight in terms of measurement values. This ordering remains consistent across both male and female groups.

7. Statistical tests

7.1. test whether gender and weight category are independent

Null: Gender and weight_category are independent (no association).
Alternative: Gender and weight_category are associated (not independent).

```python
import pandas as pd                       # tables and cross-tabs
import numpy as np                        # numeric helpers
from scipy.stats import chi2_contingency  # chi-square test

# --- 1) Clean inputs: drop rows where either variable is missing ---
df = data[['gender', 'weight_category']].copy()   # select the two categorical columns
df = df.dropna(subset=['gender', 'weight_category'])  # Chi-square needs observed categories only

# --- 2) Build contingency table (rows=gender, cols=weight_category) ---
ct = pd.crosstab(df['gender'], df['weight_category'])
print("Contingency table:")
print(ct)

# --- 3) Chi-square test of independence ---
use_correction = (ct.shape == (2, 2))
chi2, p, dof, expected = chi2_contingency(ct, correction=use_correction)

print("\nChi-square test results:")
print(f"chi2 = {chi2:.4f}, dof = {dof}, p-value = {p:.6f}")
```

* The chi-square test found a statistically significant relationship between gender and weight category (p-value = 0.000485, which is less than 0.05).
* This means we reject the null hypothesis - gender and weight category are NOT independent. They are associated with each other.
* Females: More are in the "low weight" category (17 out of 41 total) and fewer are in the "normal weight" category (21 out of 41)
* Males: Most are in the "normal weight" category (41 out of 46 total), with very few in either "high weight" (1) or "low weight" (4)

7.2. check the effect of means of body measurements and weight category

* Null: The vector of means for all body measurements is the same across all weight_category groups.
* Alternative: At least one group mean vector differs (there is an overall multivariate effect of weight_category)

```python
# MANOVA: body measurements ~ weight_category
import pandas as pd
import numpy as np
from statsmodels.multivariate.manova import MANOVA

# body-measurement columns
measures = ['ankle','arm-length','bicep','calf','chest','forearm','height','hip',
            'leg-length','shoulder-breadth','shoulder-to-crotch','thigh','waist',
            'wrist']

df = data.copy()
df['weight_category'] = df['weight_category'].astype('object')

# coerce numeric
for c in measures:
    df[c] = pd.to_numeric(df[c], errors='coerce')

used_cols = ['weight_category'] + measures
df_clean = df[used_cols].dropna(axis=0, how='any')

# quote any columns with special chars using Q("colname")
q_measures = [f'Q("{c}")' for c in measures]
formula = ' + '.join(q_measures) + ' ~ C(weight_category)'

ma = MANOVA.from_formula(formula, data=df_clean)
mv_res = ma.mv_test()
print("\nMANOVA results:")
print(mv_res)
```

* The MANOVA found a statistically significant relationship (p-value = 0.0000 for all tests, which is way less than 0.05).
* This means we reject the null hypothesis - the body measurements ARE different across weight categories. At least one weight category group has different average body measurements compared to the others.
* When looking at all body measurements together, weight category makes a significant difference. People in different weight categories (high weight, low weight, normal weight) have statistically different patterns of body measurements.
* Weight category significantly affects body measurements. 

7.3. check the effect of weight category over height

* Null: no mean difference across weight categories
* Alternative: At least one weight_category group mean for height_cm differs

```python
# One-way ANOVA: height_cm ~ weight_category
import statsmodels.api as sm                 # ANOVA
from statsmodels.formula.api import ols      # OLS for ANOVA
from statsmodels.stats.anova import anova_lm # ANOVA table

# --- 1) Prepare data ---
df = data[['height_cm', 'weight_category']].copy()           # select needed columns
df['weight_category'] = df['weight_category'].astype('object')# ensure categorical-like
df['height_cm'] = pd.to_numeric(df['height_cm'], errors='coerce')  # coerce to numeric
df = df.dropna(subset=['height_cm', 'weight_category'])      # drop rows with missing values

# Quick sanity check: need 2 groups with data
group_sizes = df['weight_category'].value_counts()
if (group_sizes >= 2).sum() < 2:
    raise ValueError("Not enough groups with 2 observations for ANOVA.")

# --- 2) Fit OLS model and run ANOVA (Type II) ---
model = ols('height_cm ~ C(weight_category)', data=df).fit()  # encode categories
aov_table = anova_lm(model, typ=2)                            # ANOVA table (SS, df, F, PR(>F))
print("ANOVA (height_cm ~ weight_category):")
print(aov_table)
```

* Since the p-value is extremely small (much less than 0.05), we reject the null hypothesis. This means there is very strong statistical evidence that average height differs significantly across the weight categories.
